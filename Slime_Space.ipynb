```python
def frame_counter(videoCapture):
    frame_count = 0
    while (ret0 == True):
        ret0, frame0 = videoCapture.read()
        if ret0 is False:
            break
        frame_count = frame_count + 1
    return frame_count

def Width(frame0):
    return frame0.shape[1]

def Height(frame0):
    return frame0.shape[0]

def CutFrame(f, a, b):
    _f = f[a:b]
    return _f

def Difference(videoCapture1, videoCapture2, number_frame):  #废物
    sum_dif = 0
    overall_space = []
    diff = []
    for m in range(number_frame - 1):
        print(m , end='\r') 
        dif = 0
        ret1, frame1 = videoCapture1.read()  # start from frame1
        ret2, frame2 = videoCapture2.read()  # start from frame0
        f1 = np.array(frame1)
        np.squeeze(f1)
        f2 = np.array(frame2)
        np.squeeze(f2)
        CutFrame(f1, 250, 750)
        CutFrame(f2, 250, 750)
        dif += np.sum(np.abs(f1 - f2))
        sum_dif += dif
        overall_space.append(sum_dif)
        diff.append(dif)
    plt.plot(diff, c='r')
    plt.plot(overall_space, c='b')
    plt.title('Summary & Differential of pixel changes')
    plt.show()
    
def show_slime_occupy(VideoCapture, number_frame, num_out): #红！
    for i in range(number_frame//num_out):
        ret, frame = VideoCapture.read()
    for i in range(0, number_frame-number_frame//num_out-1, number_frame//num_out):
        for j in range(number_frame // num_out):
            ret, frame = VideoCapture.read()
        f = frame[250:750]
        p = 0
        for x in range(500):
            for y in range(3264):
                if f[x][y][0] <= 50:#Changed here!!!
                    f[x][y] = [0, 0, 255]
                    p += 1
        print(i+number_frame//num_out, ':', p * 100 // (500 * 3264), '%')
        cv2.imwrite('%d'%i+'slime_occupy.jpg', f)
        
def Capture_Yellow(videoCapture, number_frame):  #出图
    yellow_pixel = []
    for m in range(number_frame):
        n_yellow = 0
        ret, frame = videoCapture.read()
        f = np.array(frame)
        np.squeeze(f)
        f_c = CutFrame(f, 800, 1300)
        f_b = f_c[:,:,0]
        np.squeeze(f_b)
        flag = np.ones((500, 3264))
        flag *= 50
        f_b = np.mat(f_b)
        flag = np.mat(flag)
        f_b = f_b - flag
        f_b = np.array(f_b)
        for i in range(500):
            for j in range(3264):
                if f_b[i][j] < 0:
                    n_yellow += 1
        print(m, n_yellow)
        yellow_pixel.append(n_yellow)
    #print(yellow_pixel)
    diff = []
    for i in range(len(yellow_pixel)-1):
        dif = yellow_pixel[i+1] - yellow_pixel[i]
        diff.append(dif)
        
    plt.plot(diff, c = 'r')
    plt.plot(yellow_pixel, c = 'b')
    plt.title('Summary & Differential of yellow pixel')
    plt.xlabel('Frames')
    plt.ylabel('Number of Pixels/%')
    plt.show()
    
```
